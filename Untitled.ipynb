```python
import pandas as pd

df = pd.read_csv('dccelectricitycivicsblocks34p20130221-1841.csv', engine='python', skipfooter=3)
df
```

```python
# Data is sorted on a daily basis instead of 15 mins
import numpy as np
import matplotlib.pyplot as plt

df2=df.rename(columns=df.iloc[0])
df3=df2.drop(df.index[0])
df3
df3.drop(df3.index[0])
df4=df3.drop('Date', axis=1)
df5=df4.drop('Values', axis=1)
df5
df6=df5.dropna()
df7=df6.values
df7
dataset=np.sum(df7, axis=1, dtype=float)
dataset
plt.plot(dataset)
plt.xlabel('Days')
plt.ylabel('Kilowatts')
plt.title("Kilowatts consumed per day")
plt.show()
```

```python
#analyze the data set

from statsmodels.tsa.seasonal import seasonal_decompose

decomposition=seasonal_decompose(dataset, model='additive', period=1)
trend=decomposition.trend
seasonal=decomposition.seasonal
residual=decomposition.resid
plt.subplot(221)
plt.plot(dataset,color='#ff0000', label='Series')
plt.legend(loc='best')
plt.subplot(222)
plt.plot(trend,color='#1100ff', label='Trend')
plt.legend(loc='best')
plt.tight_layout()
plt.subplot(223)
plt.plot(residual,color='#00ff1a', label='Residual')
plt.legend(loc='best')
plt.tight_layout()
plt.subplot(224)
plt.plot(seasonal,color='#de00ff', label='Seasonality')
plt.legend(loc='best')
plt.tight_layout()
plt.show()
```

```python
# The first 150 days used for calculations
train_df=dataset[:150]
train_df
```

```python
# Manual ARIMA configuration

from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Partial Autocorrelation

plot_pacf(train_df, lags=10, zero=False);

# as oartial autocorrelation shows a sudden drop at lag 2, p=2 is specified.
```

```python
# Autocorrelation of differenced series - trends and seasonality components removed

plot_acf(train_df, lags=10, zero=False);

# There is no significant positive correlation for the first few lags. so q=0.

# Autocorrelation function showed evidence of stationarity by first-differencing the series so d=1
```

```python
test_df=dataset[150:208]
```

```python
window_size = 7

numbers_series = pd.Series(train_df)
windows = numbers_series.rolling(window_size)
moving_averages = windows.mean()

moving_averages_list = moving_averages.tolist()
ts7 = moving_averages_list[window_size - 1:]
ts7=np.array(ts7)
print(ts7)
```

```python
import math

n1=math.nan
n7=np.array([n1,n1,n1,n1,n1,n1,n1])
ts7=np.concatenate([n7,ts7])
plt.plot(train_df)
plt.plot(ts7)
plt.xlabel('Days')
plt.ylabel('Kilowatts')
plt.title("7-day Simple Moving Average")
plt.show()
```

```python
#Dickey-Fuller Test
import statsmodels.tsa.stattools as ts
result = ts.adfuller(train_df, 1)
result
```
