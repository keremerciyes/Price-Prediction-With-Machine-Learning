```python
import tweetnlp
import pandas as pd
import torch.utils.data as data_utils
```

```python
data = pd.read_csv("bitcoin_tweets_only_verified_with_date_and_tweets.csv")
```

```python
model = tweetnlp.load_model("sentiment")
output = model.sentiment(data["text"], batch_size=1000)
```

```python
data_with_sentiments = data
data_with_sentiments['sentiment'] = [item['label'] for item in output]
data_with_sentiments.head(25)
```

```python
data_with_sentiments.to_csv("bitcoin_tweets_only_verified_with_date_tweets_and_roBERTa_sentiments_.csv", index=False)
```
